```python
sns.countplot(data=df, x='type')
plt.title('Distribucion de peliculas y series')
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', title='Top de 10 pases con ms ttulos')
plt.xlabel('Cantidad de titulos')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(12,4), title='Titulos por ao de lanzamiento')
plt.show()
```
